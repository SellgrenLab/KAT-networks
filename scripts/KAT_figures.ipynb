```python
import anndata as ad
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scanpy as sc
```

```python
gene_name="GOT2"
gene_data = pd.DataFrame(adata[:, gene_name].X.toarray(), columns=[gene_name], index=adata.obs.index)
gene_data["Time Point"] = adata.obs["agegroup"]
gene_data["region"] = adata.obs["broad_region"]
ax=sns.lineplot(data=gene_data, x="Time Point", y=gene_name, hue="region")
# Add data points
#sns.scatterplot(data=gene_data, x="Time Point", y=gene_name, color='r', s=10)
# set y-axis limits
ax.set(ylim=(0, 0.8))
plt.title(f"{gene_name} Expression Across Time Points")
plt.xlabel("Developmental Period")
plt.ylabel("GOT2 Expression")
plt.show()

# save the plot
plt.savefig("GOT2_fetal_agegroup.pdf")
```
